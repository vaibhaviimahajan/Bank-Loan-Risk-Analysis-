```python
# Fill missing employment titles
df['emp_title'] = df['emp_title'].fillna("Unknown")
```

```python
df
```

```python
#Check Data types
df['emp_length'].dtype
df['term'].dtype
```

```python
# Convert employment length to numeric
df['emp_length'] = (
    df['emp_length']
    .astype(str)
    .str.extract(r'(\d+)')
    .astype(float)
)
```

```python
# Convert loan term to numeric months
df['term'] = (
    df['term']
    .astype(str)
    .str.extract(r'(\d+)')
    .astype(int)
)
```

```python
# Create loan performance flag
df['loan_flag'] = np.where(
    df['loan_status'] == 'Charged Off',
    'Bad Loan',
    'Good Loan'
)
     
```

```python
# Extract issue year-month for trend analysis
df['issue_month'] = df['issue_date'].dt.to_period('M')
```
